```python
%load_ext autoreload
%autoreload 2
import numpy as np
from MachineLearning.IsoMap.isomap import isomap, delta_matrix, shortest_pathFW
from MachineLearning.utils.distance import matrixEDM
from scipy.sparse.csgraph import shortest_path
from MachineLearning.MDS.mds import mds, mds2
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
from mpl_toolkits.mplot3d import Axes3D
from MachineLearning.KNearestNeighbor.knn import kNearestNeighborSearch, vectorizedNearestNeighbor
from sklearn.neighbors import kneighbors_graph
from sklearn import manifold
```

```python
X_raw = np.loadtxt('MachineLearning/datasets/3Ddata.txt')
X = X_raw[:, :3]
Y = X_raw[:, 3]
```

```python
fig = plt.figure(figsize=(10,8))
#ax = fig.add_subplot(111, projection='3d')
ax = Axes3D(fig, elev=18, azim=100)
ax.scatter(X.T[0],X.T[1],X.T[2], c=Y, cmap='Vega20b',s=50)
plt.show()
```

## MDS

```python
y = mds(X, 2)
```

```python
y.shape
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.scatter(y.T[0], y.T[1], c = Y, cmap='Vega20b')
plt.show()
```

```python
scikitMDS = manifold.MDS(n_components=2, dissimilarity='euclidean')
y = scikitMDS.fit_transform(X)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.scatter(y.T[0], y.T[1], c = Y, cmap='Vega20b')
plt.show()
```

## IsoMap

```python
y = isomap(X, 15, 2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y.T[0], y.T[1], c=Y, cmap='Vega20b')
plt.show()
```

```python
y = manifold.Isomap(n_neighbors=15, n_components=2).fit_transform(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y.T[0], y.T[1], c=Y, cmap='Vega20b')
plt.show()
```
